```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
city_data_df = pd.read_csv('Pyber/raw_data/city_data.csv')
ride_data_df = pd.read_csv('Pyber/raw_data/ride_data.csv')
ride_data_with_city_type = pd.merge(ride_data_df, city_data_df, on='city', how='left')
ride_data_with_city_type = ride_data_with_city_type.drop('driver_count', axis=1)

# print(city_data_df.info())
print(city_data_df.head())
print()
print(ride_data_df.head())
print()
ride_data_with_city_type
```

```python
average_fare_per_city = (ride_data_df.groupby('city'))['fare'].sum() / (ride_data_df.groupby('city'))['fare'].count()
rides_per_city = (ride_data_df.groupby('city'))['ride_id'].count()
drivers_per_city = (city_data_df.groupby('city'))['driver_count'].sum()
city_types = city_data_df['type'].unique()

print(average_fare_per_city)
print()
print(rides_per_city)
print()
print(drivers_per_city)
print()
print(city_types)
```

```python
percentage_fare_by_city_type = (ride_data_with_city_type.groupby('type'))['fare'].sum() / ride_data_with_city_type['fare'].sum()
percentage_ride_by_city_type = (ride_data_with_city_type.groupby('type'))['ride_id'].count() / ride_data_with_city_type['ride_id'].count()
percentage_driver_by_city_type = (city_data_df.groupby('type'))['driver_count'].sum() / city_data_df['driver_count'].sum()

print(percentage_fare_by_city_type)
print()
print(percentage_ride_by_city_type)
print()
print(percentage_driver_by_city_type)
print()
```
